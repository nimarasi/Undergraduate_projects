```python
#%%
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
#read excel file
df = pd.read_excel("Results.xlsx")
df
```

```python
# split part 1 & 2 & 3 questions
p1 = df.iloc[:,5:105]
p2 = df.iloc[:, 105:125]
p3 = df.iloc[:,125:]
```

```python
p1.describe()
p3.describe()
```

```python
# feature scaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
p1 = pd.DataFrame(scaler.fit_transform(p1),columns=p1.columns)
p3 = pd.DataFrame(scaler.fit_transform(p3),columns=p3.columns)
```

```python

fig, ax = plt.subplots(1,3,figsize=(40,300))
plt.figure(dpi=250)
fig.tight_layout()

# df["Gender"].value_counts()
b = sum(np.array(df["Gender"] == 0))
g = sum(np.array(df["Gender"] == 1))
pie_chart_x = np.array([b,g])
label = ["male" , "female"]
ax[0].pie(pie_chart_x, labels=label, autopct='%0.f%%', shadow=True, textprops={'fontsize': 28});
ax[0].set_title('Gender pie plot',fontsize=28)


# df["City"].value_counts()

city_dict = {"city_300" : 17 , "city_900" : 12 , "city_1700" : 2 , "city_600" : 1 ,"city_1125" : 1 , "city_430" : 1
             , "city_350" : 1, "city_1325" : 1, "city_1500" : 1, "city_800" : 1, "city_500" : 1, "city_1200" : 1}

city_names = np.array(list(city_dict.keys()))
city_values = np.array(list(city_dict.values()))

ax[1].pie(city_values, labels=city_names, autopct='%0.f%%', shadow=True,  textprops={'fontsize': 28});
ax[1].set_title('City pie plot',fontsize=28)


month_dict = dict(df['Month'].value_counts())

month_names = np.array(list(month_dict.keys()))
month_values = np.array(list(month_dict.values()))
label=list(set(df["Month"]))
ax[2].pie(month_values, labels=month_names, autopct='%0.f%%', shadow=True, textprops={'fontsize': 28});
ax[2].set_title('Month pie plot',fontsize=28)
```

```python
df_normalize = pd.concat([p1,p2,p3], axis=1)
df_normalize.index = df.iloc[:,0]
df_normalize
```

```python
distance_df = pd.DataFrame(np.zeros((30,30)), index= df.iloc[:,0] ,columns=df.iloc[:,0])
distance_df
```

```python
from scipy.spatial import distance
```

```python
for i in range(0,30):
    for j in range(0,30):
      distance_df.iloc[i,j] = distance.euclidean(df_normalize.iloc[i,:], df_normalize.iloc[j,:])

distance_df
```

```python
min_max_dist = pd.DataFrame(columns=["Min","Min_User","Max", "Max_User"],index=df.iloc[:,0])

for i in range(30):
  temp = np.array(distance_df.iloc[i,:])
  temp = temp[temp != 0]
  min_max_dist.iloc[i,0] = temp.min()
  min_max_dist.iloc[i,1] = distance_df[distance_df.iloc[i,:] == temp.min()].index[0]
  min_max_dist.iloc[i,2] = temp.max()
  min_max_dist.iloc[i,3] = distance_df[distance_df.iloc[i,:] == temp.max()].index[0]
```

```python
min_max_dist
```

```python
plt.figure(figsize=(40, 30))
sns.heatmap(distance_df.corr().abs(),  annot=True);
```

```python
import scipy.cluster.hierarchy as shc


plt.figure(figsize=(24, 16))
plt.title("hierarchy clustering Dendrogram")
plt.grid(False)

# Selecting Annual Income and Spending Scores by index
clusters = shc.linkage(df_normalize,
            method='ward',
            metric="euclidean")
shc.dendrogram(Z=clusters)
plt.xlabel('Users')
plt.ylabel('distance')
plt.axhline(y=7, color='r', linestyle='--')

plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')
cluster.fit_predict(df_normalize)
```

```python
cluster_df = pd.concat([df.iloc[:,0] , pd.DataFrame(cluster.fit_predict(df_normalize), columns=["Cluster"])] ,axis=1)
```

```python
cluster_df
```

```python
c0 = cluster_df[cluster_df.iloc[:,1] == 0]
c1 = cluster_df[cluster_df.iloc[:,1] == 1]
c2 = cluster_df[cluster_df.iloc[:,1] == 2]
c3 = cluster_df[cluster_df.iloc[:,1] == 3]
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(df_normalize["Q60"], df_normalize['Q54'], c=cluster.labels_)
```

```python
import matplotlib.patches as mpatches
plt.figure(figsize=(10,8) , dpi=150)


x = df["Month"]
y = distance_df
markers = [ "s", "d", "*", "^"]

for xp, yp in zip(x, y):
  if np.any(c0.isin([yp])):
    plt.scatter(xp, yp, marker="s", color="r", s=100)
  elif np.any(c1.isin([yp])):
        plt.scatter(xp, yp, marker="d" , color="b", s=100)
  elif np.any(c2.isin([yp])):
        plt.scatter(xp, yp, marker="*" , color="k", s=100)
  elif np.any(c3.isin([yp])):
    plt.scatter(xp, yp, marker="^", color="g",s=100)

plt.xticks(x)
plt.xlabel("Month")

plt.ylabel("Users")

red_patch = mpatches.Patch(color='red', label='Cluster0')
blue_patch = mpatches.Patch(color='blue', label='Cluster1')
black_patch = mpatches.Patch(color='black', label='Cluster2')
green_patch = mpatches.Patch(color='green', label='Cluster3')


plt.legend(handles=[red_patch , blue_patch,black_patch , green_patch])
plt.show()
```

```python
distance_df_binary = pd.DataFrame(index= df.iloc[:,0] ,columns=df.iloc[:,0])
for i in range(30):
  distance_df_binary.iloc[i] = distance_df.iloc[i].apply(lambda x: 1 if x > distance_df.iloc[i].median() else 0)
distance_df_binary
```

```python
import networkx as nx

G = nx.Graph()
L=[]
for i in range(30):
  for j in range(30):
    if distance_df_binary.iloc[i,j] == 1:
      L.append(('User{}'.format(i+1),'User{}'.format(j+1)))

G.add_edges_from(L)

plt.figure(figsize=(10,8) , dpi=150)
nx.draw_networkx(G, with_labels = True, node_color ='green',font_color="b")
```
